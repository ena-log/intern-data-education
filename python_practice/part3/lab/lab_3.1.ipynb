# beautiful soup library를 이용하여 파싱

```python
from bs4 import BeautifulSoup
```

```python
naver_url="https://movie.naver.com/movie/point/af/list.nhn" 
```

```python
from urllib.request import urlopen 
```

```python
response = urlopen(naver_url)
encoding = response.info() .get_content_charset(failobj="utf-8")
html = response.read().decode(encoding)
a = open("data/naver_movie.html", "w")
a.write(html)
a.close
```

```python
with open('data/naver_movie.html') as f:
    soup = BeautifulSoup(f, 'html.parser')
```

```python
import re
```

```python
for x in soup.find_all('td'):
    print(x.text)
```

```python
for x in soup.find_all('td'):
    content = re.search("<br/>[ 가-힣]+", str(x)) 
    if content == None :
        continue;
    af = x.find_all('a') 
    for idx , detail in enumerate( af )   :
        print("영화제목 = ", detail.text) 
        print("영화평=" , str(content)[53:-2])
        print()
        break;
```

 # Ixml 라이브러리를 사용하여 파싱

```python
import lxml.html
```

```python
tree = lxml.html.parse('data/naver_movie.html')
```

```python
html = tree.getroot()
```

```python
for a in html.cssselect('a'):
    print(a.text)
```
