```python
from selenium import webdriver

front_url = "https://section.blog.naver.com/Search/Post.nhn?pageNo="
back_url = "&rangeType=ALL&orderBy=sim&keyword="
place = "서울명소"

allurl = []
for i in range(0,250):
    if i%10==1 :
        url = front_url + str(i) +  back_url + place
        allurl.append(url)
                
driver = webdriver.Chrome('driver/chromedriver')

for i in range(0,25):
    driver.get(allurl[i])

    print(len(driver.window_handles))

    from bs4 import BeautifulSoup

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    result = []
    for span in driver.find_elements_by_css_selector('strong > span.title'):
         result.append(span.text)
            
    #print(result)
    
    all = " "
    for i in result :
        all = all +i 
        all = all.replace('1','')
    print(all)
 
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

 
font_path = 'c:/Windows/fonts/malgun.ttf'
wordcloud = WordCloud(
     font_path = font_path,
    width = 800,
    height = 800
    )

wordcloud = wordcloud.generate(all)
 
fig = plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
fig.savefig('wordcloud_without_axisoff.png')

```
