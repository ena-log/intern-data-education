```python
import math
import time
import csv
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome('driver/chromedriver')

qList = []

f = open('실내여행지(전국).csv', 'r', encoding='euc-kr')
rdr = csv.reader(f)
for line in rdr:
#     print(line[1])
    qList.append(line[1])
f.close()

# print(qList)

date1 = ['20200101', '20200201', '20200301', '20200401', '20200501', '20200601']
date2 = ['20200131', '20200229', '20200331', '20200430', '20200531', '20200630']

data = []

for qIdx, query in enumerate(qList):

    print(str(qIdx+1) + "/" + str(len(qList)))


    for dIdx in range(0, len(date1)):

        url = "https://search.naver.com/search.naver?where=post&query=" + query + "&st=sim&sm=tab_opt&date_from=" + date1[dIdx] + "&date_to=" + date2[dIdx] + "&date_option=8&srchby=all&dup_remove=1&nso=so%3Ar%2Ca%3Aall%2Cp%3Afrom" + date1[dIdx] + "to" + date2[dIdx]
        driver.get(url)

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        try:
            title_num = soup.find('span', class_="title_num")
            title_num = title_num.text
            cut = title_num.index('/')
            title_num = title_num[cut+2:-1]
            title_num = title_num.replace(",", "")
            title_num = int(title_num)
#             print(title_num)

            data.append([query, dIdx+1, title_num])
        except:
            print("error pass")
#             print(data)

        time.sleep(1)

csvName = "output2.csv"
f = open(csvName, 'w', encoding='euc-kr', newline='')
wr = csv.writer(f)
for idx, item in enumerate(data):
    wr.writerow(item)
f.close()

print("완료")
```
