```python
import pandas as pd
```

```python
#2019년 보행노인 사고다발 지역 데이터를 csv로 불러옴.
ac_2019=pd.read_csv('보행노인 사고다발지역정보 데이터셋_동_2019.csv', encoding = 'cp949', header=0)
```

```python
ac_2019
```

```python
ac_2019_1=ac_2019[['장소설명','사고건수','사상자수','사망자수','중상자수','경상자수','부상신고자수']]
#불러온 csv에서 필요한 column 들만 남겨서 새로운 list 생성
```

```python
ac_2019_1
```

```python
#2013~2020년 까지 사고내역 csv를 불러옴
ac_2013_2020=pd.read_csv('보행노인 사고다발지역정보_동_2013-2020.csv', encoding='cp949')
```

```python
ac_2013_2020
```

```python
#column명을 처음 불러온 csv에 맞춤.
ac_2013_2020.rename(columns={"지점명":"장소설명"}, inplace=True)
```

```python
#column명을 처음불러온 csv에 맞춤
ac_2013_2020.rename(columns={'발생건수':'사고건수'},inplace=True)
```

```python
ac_2013_2020
```

```python
ac_total=ac_2013_2020[['장소설명','사고건수','사상자수','사망자수','중상자수','경상자수','부상신고자수','경도','위도']]
#필요한 column들만 남겨서 새로운 list 생성
```

```python
ac_total
```

```python
ac_2019_1[['장소설명','사고건수','사상자수','사망자수','중상자수','경상자수','부상신고자수']].astype(str)
#merge를 위해 string 타입으로 column 설정
```

```python
ac_total[['장소설명','사고건수','사상자수','사망자수','중상자수','경상자수','부상신고자수']].astype(str)
#merge를 위해 string 타입으로 column 설정
```

```python
# key 값으로  ['장소설명','사고건수','사상자수','사망자수','중상자수','경상자수','부상신고자수'] 사용해서 inner join 
merge_ac_2019=pd.merge(ac_2019_1,ac_total, on=['장소설명','사고건수','사상자수','사망자수','중상자수','경상자수','부상신고자수'], indicator=True)
```

```python
merge_ac_2019
```

```python
 merge_ac_2019.to_csv("사고다발지역 위경도.csv", encoding='cp949')
# merge 한 데이터 셋을 csv로 저장
```
